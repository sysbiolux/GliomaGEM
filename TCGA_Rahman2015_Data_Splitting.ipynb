```python
import pandas as pd
```

```python
filename = 'data/TCGA Rahman, et al 2015/GSM1697009_06_01_15_TCGA_24.normal_Rsubread_FPKM.txt'
```

```python
meta = pd.read_csv('data/TCGA_TCGBiolinks_metadata_Summary.csv',index_col=0)
```

```python
meta.head()
```

## Read the control glioma samples

```python
import dask.dataframe as dd
df_ctrl = dd.read_csv(filename,sep='\t').set_index('Unnamed: 0')
```

```python
df_ctrl.shape
```

```python
df_ctrl.head()
```

```python
## Control samples in LGG and GBM
glioma_ctrl_samples = [x for x in df_ctrl.columns if x in meta.index]
glioma_ctrl = df_ctrl.loc[:,glioma_ctrl_samples]
```

```python
glioma_ctrl_samples
```

```python
glioma_ctrl.head()
```

## Read the tumor glioma samples

```python
import dask.dataframe as dd
filename = 'data/TCGA Rahman, et al 2015/GSM1536837_06_01_15_TCGA_24.tumor_Rsubread_FPKM.txt'
df_tumor = dd.read_csv(filename,sep='\t', sample=1000000).set_index('Unnamed: 0')
```

```python
df_tumor.shape
```

```python
df_tumor.head()
```

```python
## Control samples in LGG and GBM
glioma_tumor_samples = [x for x in df_tumor.columns if x in meta.index]
glioma_tumor = df_tumor.loc[:,glioma_tumor_samples]
```

```python
glioma_tumor.head()
```

```python
glioma_tumor.shape
```

```python
glioma_df = dd.merge(glioma_tumor, glioma_ctrl, on=['Unnamed: 0'])
```

```python
glioma_df.head()
```

```python
glioma_df.to_csv('data/TCGA Rahman, et al 2015/TCGA_LGG_GBM.csv', single_file = True)
```
